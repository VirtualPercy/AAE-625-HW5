```python
import os
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load ""ManagingABank_AAE625 - Sheet1.csv""
df = pd.read_csv("ManagingABank_AAE625 - Sheet1.csv")
# # convert the "NaN" value in "Team 1 Bid", ""Team 2 Bid", "Team 3 Bid" to 0%
df["Team 1 Bid"] = df["Team 1 Bid"].fillna(0)
df["Team 2 Bid"] = df["Team 2 Bid"].fillna(0)
df["Team 3 Bid"] = df["Team 3 Bid"].fillna(0)
# df["Team 4 Bid"] = df["Team 4 Bid"].fillna(0)
df.head(11)
```

```python
# for each row in the dataframe, if the value in "Team 1 Bid" is not 0,
# check if the value in "Team 1 Bid" is greater than or equal to the value in "Team 2 Bid",
# if so, set the value in "Team 1 Bid" equal to the value in "Team 2 Bid" - 0.0025.
# stop when the value in "Team 1 Bid" is lower than or equal to 0.03

for index, row in df.iterrows():
    if row["Team 1 Bid"] != 0:
        if row["Team 2 Bid"] > row["Team 3 Bid"]:
            if row["Team 1 Bid"] > row["Team 3 Bid"] and row["Team 3 Bid"] > 0.03:
                df.at[index, "Team 1 Bid"] = row["Team 3 Bid"] - 0.0025
            elif row["Team 1 Bid"] == row["Team 2 Bid"]:
                df.at[index, "Team 1 Bid"] = row["Team 2 Bid"] - 0.0025
        elif row["Team 2 Bid"] <= row["Team 3 Bid"]:
            if row["Team 1 Bid"] > row["Team 2 Bid"] and row["Team 2 Bid"] > 0.03:
                df.at[index, "Team 1 Bid"] = row["Team 2 Bid"] - 0.0025
            elif row["Team 1 Bid"] == row["Team 3 Bid"]:
                df.at[index, "Team 1 Bid"] = row["Team 3 Bid"] - 0.0025
        

df.head(11)

```

```python
# change the value in "Team 1 Bid" to "NaN" if the value in "Team 1 Bid" is 0
df["Team 1 Bid"] = df["Team 1 Bid"].replace(0, np.nan)
```
